Problem Statement


HELP International is an international humanitarian NGO that is committed to fighting poverty and providing the people of backward countries with basic amenities and relief during the time of disasters and natural calamities. It runs a lot of operational projects from time to time along with advocacy drives to raise awareness as well as for funding purposes.

After the recent project that included a lot of awareness drives and funding programmes, they have been able to raise around $ 10 million. Now the CEO of the NGO needs to decide how to use this money strategically and effectively. The significant issues that come while making this decision are mostly related to choosing the countries that are in the direst need of aid.

And this is where you come in as a data analyst. Your job is to categorise the countries using some socio-economic and health factors that determine the overall development of the country. Then you need to suggest the countries which the CEO needs to focus on the most.

# EDA

```python
# Lowest 10 countries for each factor:

# The child mortality rate (also under-five mortality rate) refers to the probability of dying between birth and
# exactly five years of age expressed per 1,000 live births.

top10_child_mortality = df[['country','child_mort']].sort_values('child_mort',ascending=False).head(10)
```

```python
# Exports are goods and services that are produced in one country and sold to buyers in another.

top10_exports = df[['country','exports']].sort_values('exports',ascending=True).head(10)
```

```python
# . Health status is a measure of how people perceive their health—rating it as excellent, very good, good, fair, or poor
# . Reported health status is a predictor of important health outcomes including 
# mortality, morbidity, and functional status (1–4).
```

```python
health_based_cont = df[['country','health']].sort_values('health',ascending= True).head(10)
```

```python
# An import is a good or service bought in one country that was produced in another. 

Top10_imports = df[['country','imports']].sort_values('imports',ascending= True).head(10)
```

```python
# Gross National Income (GNI) is the total amount of money earned by a nation's people and businesses.
# It is used to measure and track a nation's wealth from year to year.

top10_income = df[['country','income']].sort_values('income',ascending=True).head(10)
```

```python
# Inflation is typically a broad measure, such as the overall increase in prices or the increase
# in the cost of living in a country.

top10_inflation = df[['country','inflation']].sort_values('inflation',ascending=True).head(10)
```

```python
# Life Expectancy: The average number of years a new born child would live if the current mortality patterns are to remain same

top10_life_expec = df[['country','life_expec']].sort_values('life_expec',ascending=True).head(10)
```

```python
# Fertility Rate: The number of children that would be bornto each woman if the current age-fertility rates remain the same

top10_fertility_rate = df[['country','total_fer']].sort_values('total_fer',ascending=True).head(10)
```

```python
# The GDP per capita : Calculated as the Total GDP divided by the total population.

# Fertility Rate: The number of children that would be bornto each woman if the current age-fertility rates remain the same

top10_GDPP = df[['country','gdpp']].sort_values('gdpp',ascending=True).head(10)
```

# PCA ON THE DATA

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(svd_solver= 'randomized' , random_state = 17 )
```

```python
x = data.drop('country',axis = 1)
y = data['country']
```

```python
pca.fit(x)
```

```python
pca.components_
```

```python
pca.components_.shape
```

```python
pca.explained_variance_
```

```python
pca.explained_variance_ratio_
```

```python
column_names = list(x.columns)
```

```python
pca_data = pd.DataFrame({'Pca1':pca.components_[0],'pca2':pca.components_[1],'features':column_names})
pca_data.head()
```

```python
fig = plt.figure(figsize = (9,7))

plt.scatter(pca_data.Pca1, pca_data.pca2)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')

for i, txt in enumerate(pca_data.features):
    plt.annotate(txt, (pca_data.Pca1[i],pca_data.pca2[i]))
    print(i)
    print(txt)
plt.tight_layout()
plt.show()
```

```python
pca.explained_variance_ratio_
```

```python
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('explained varaince ration')
plt.savefig('pca_no')
plt.show()
```

#Looks like 4 components are enough to describe 95% of the variance in the dataset


#We'll choose 4 components for our modeling

```python
from sklearn.decomposition import IncrementalPCA
pca_final = IncrementalPCA(n_components=4)
```

```python
df_pca = pca_final.fit_transform(x)
df_pca.shape
```

```python
df_pca = pd.DataFrame(df_pca)
df_pca.head()
```

```python
corrmat = np.corrcoef(df_pca.transpose())
corrmat
```

```python
sns.heatmap(corrmat,annot=True)
```

```python
df_pca.columns = ['PCA_1','PCA_2','PCA_3','PCA_4']
```

```python
# k means clustering
```

```python
from sklearn.cluster import KMeans

for i in range(1,7):
    km=KMeans(n_clusters=i)
    km.fit(df_pca)
    print("Cluster :",i)
    l=km.labels_
    sns.scatterplot(df_pca["PCA_1"],df_pca["PCA_2"],hue=l)
    plt.show()
```

```python
# applying agglomerative clustering and visualising

from sklearn.cluster import AgglomerativeClustering
for i in range(1,7):
    ag=AgglomerativeClustering(n_clusters=i)
    ag.fit(df_pca)
    print("Cluster :",i)
    l=ag.labels_
    sns.scatterplot(df_pca["PCA_1"],df_pca["PCA_2"],hue=l)
    plt.show()
```

```python
# hence here k means clustering is not done 
```

```python
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans
from scipy.cluster.hierarchy import linkage
from scipy.cluster.hierarchy import dendrogram
from scipy.cluster.hierarchy import cut_tree
```

```python
# i)inertia

err=[]
for i in range(1,7):
    km=KMeans(n_clusters=i)
    km.fit(df_pca)
    err.append(km.inertia_)
err
```

```python
# silhouette score

di={}
for i in range(2,7):
    km=KMeans(n_clusters=i)
    km.fit(df_pca)
    di[i]=silhouette_score(df_pca,km.labels_)
di
```

# hierarchical clustering

```python
# aglomerative clustering
```

```python
mergings = linkage(df_pca, method = "complete", metric='euclidean')
dendrogram(mergings)
plt.show()
```

```python
clusterCut = pd.Series(cut_tree(mergings, n_clusters = 5).reshape(-1,))
df_pca_hc = pd.concat([df_pca, clusterCut], axis=1)
df_pca_hc.columns = ["PC1","PC2","PC3","PC4","ClusterID"]
df_pca_hc.head()
```

```python
pca_cluster_hc = pd.concat([data['country'],df_pca_hc], axis=1, join='outer')
pca_cluster_hc.head()
```

```python
df
```

```python
clustered_data_hc = pca_cluster_hc[['country','ClusterID']].merge(df,on='country')
```

```python
clustered_data_hc.head()
```

```python
hc_clusters_child_mort = pd.DataFrame(clustered_data_hc.groupby(["ClusterID"]).child_mort.mean())
hc_clusters_exports = pd.DataFrame(clustered_data_hc.groupby(["ClusterID"]).exports.mean())
hc_clusters_health = pd.DataFrame(clustered_data_hc.groupby(["ClusterID"]).health.mean())
hc_clusters_imports = pd.DataFrame(clustered_data_hc.groupby(["ClusterID"]).imports.mean())
hc_clusters_income = pd.DataFrame(clustered_data_hc.groupby(["ClusterID"]).income.mean())
hc_clusters_inflation = pd.DataFrame(clustered_data_hc.groupby(["ClusterID"]).inflation.mean())
hc_clusters_life_expec = pd.DataFrame(clustered_data_hc.groupby(["ClusterID"]).life_expec.mean())
hc_clusters_total_fer = pd.DataFrame(clustered_data_hc.groupby(["ClusterID"]).total_fer.mean())
hc_clusters_gdpp = pd.DataFrame(clustered_data_hc.groupby(["ClusterID"]).gdpp.mean())
```

```python
df = pd.concat([pd.Series(list(range(0,5))), hc_clusters_child_mort,hc_clusters_exports, hc_clusters_health, hc_clusters_imports,
               hc_clusters_income, hc_clusters_inflation, hc_clusters_life_expec,hc_clusters_total_fer,hc_clusters_gdpp], axis=1)
df.columns = ["ClusterID", "child_mort_mean", "exports_mean", "health_mean", "imports_mean", "income_mean", "inflation_mean",
               "life_expec_mean", "total_fer_mean", "gdpp_mean"]
df
```
